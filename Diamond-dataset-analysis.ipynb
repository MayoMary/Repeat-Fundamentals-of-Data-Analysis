## Plotting Histogram¶
To visualize the numerical distribution of the data.

# Aim:

```python
from matplotlib import pyplot as plot
import seaborn as s
diamonds.hist(figsize=(15, 15))
plot.show()
```

Analysis: What determines the price?
Plotting each attribute vs price color, clarity, cut seem to have no effect on price

But

Carat affects price. It can be observed that increase in carat increases the price.

```python
plot.figure(figsize=(5,5))
plot.scatter(diamonds['color'], diamonds['price'],color='g', alpha=0.5, label="Color vs Price")
plot.xlabel("Color")
plot.ylabel("Price")
plot.legend(loc='upper left')
plot.show()
```

```python
plot.figure(figsize=(5,5))
plot.scatter(diamonds['clarity'], diamonds['price'],color='b', alpha=0.5, label="Clarity vs Price")
plot.xlabel("Clarity")
plot.ylabel("Price")
plot.legend(loc='upper left')
plot.show()
# Standard distribution
```

```python
plot.figure(figsize=(5,5))
plot.scatter(diamonds['cut'], diamonds['price'],color='m', alpha=0.5, label="Cut vs Price")
plot.xlabel("Cut")
plot.ylabel("Price")
plot.legend(loc='upper left')
plot.show()
```

```python
# Carat vs Price 
# Directly Proportional
plot.figure(figsize=(10,10))
plot.scatter(diamonds['carat'], diamonds['price'],color='r', alpha=0.5, label="Carat vs Price")
plot.xlabel("Carat")
plot.ylabel("Price")
plot.legend(loc='upper left')
plot.show()
```
